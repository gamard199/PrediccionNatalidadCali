```python
import matplotlib.pyplot as plt

agno = ['03', '04', '05', '06', '07', '08', '09', '10', '11', '12', '13', '14', '15', '16', '17']
nac_norte = [301, 333, 318, 333, 320, 351, 308, 269, 298, 288, 280, 297, 276, 288, 269]
nac_tradicional = [306, 270, 265, 258, 260, 234, 256, 237, 203, 253, 259, 228, 223, 230, 219]
nac_oeste = [177, 151, 172, 140, 189, 175, 160, 159, 159, 151, 144, 162, 128, 175, 173]

plt.plot(agno, nac_norte, color = 'blue', linewidth = 1, label = 'Nacimientos zona Norte')
plt.plot(agno, nac_tradicional, color = 'red', linewidth = 1, label = 'Nacimientos zona Tradicional')
plt.plot(agno, nac_oeste, color = 'green', linewidth = 1, label = 'Nacimientos zona Oeste')
plt.xlabel('Año')
plt.ylabel('Número de nacimientos por año')
plt.title('Grafica de tendencia de indice de natalidad por zonas por año')
plt.legend()
plt.show()
```
