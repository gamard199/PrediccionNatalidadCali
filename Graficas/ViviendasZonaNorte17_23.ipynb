```python
import matplotlib.pyplot as plt

agno = ['2017', '2022', '2023']
SenderosFlora = [5666, 6011, 7128]
UrbFlora  = [3914, 4661, 5295]
CiudadAlamos = [1665, 1665, 1665]
Menga = [547, 678, 808]
ElBosque = [1081, 1081, 1420]

plt.plot(agno, SenderosFlora, color = 'blue', linewidth = 1, label = 'Senderos de La Flora')
plt.plot(agno, UrbFlora, color = 'red', linewidth = 1, label = 'Urbanización La Flora')
plt.plot(agno, CiudadAlamos, color = 'purple', linewidth = 1, label = 'Ciudad de Los Alamos')
plt.plot(agno, Menga, color = 'black', linewidth = 1, label = 'Menga')
plt.plot(agno, ElBosque, color = 'green', linewidth = 1, label = 'El Bosque')
plt.xlabel('Año')
plt.ylabel('Número de viviendas')
plt.title('Número de viviendas registradas por barrio para los años 2018 a 2023')
plt.legend()
plt.show()
```
